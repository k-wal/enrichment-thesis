```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os
from statistics import mean
sns.set_theme()
sns.set(font_scale=1.5) 
sns.set_style('darkgrid')
```

```python
plt.rcParams["figure.figsize"] = (14,10)
plt.rcParams.update({'font.size': 22})
```

```python
def eda(filepath):
    big_ents = 0
    big_ent_dens = 0
    df = pd.read_csv(filepath)
    sent_lens = []
    ents_per_sent = []
    ent_lens = []
    all_ents = []
    big_sents = 0
    counts = {'B':0, 'I':0, 'O':0}
    
    for index, row in df.iterrows():
        words = row['sentence'].split(' ')
        tags = row['word_labels'].split(',')
        sent_lens.append(len(words))
        
        if len(words) > 120:
            big_sents += 1
        
        ents = []
        cur_ent = ''
        for tag in tags:
            if tag == 'O':
                counts['O'] += 1
                if cur_ent != '':
                    ents.append(cur_ent)
                    cur_ent = ''
                continue
            if tag == 'B':
                counts['B'] += 1
                if cur_ent != '':
                    ents.append(cur_ent)
                cur_ent = 'B'
            if tag == 'I':
                counts['I'] += 1
                cur_ent += 'I'
        
        if len(ents) > 10:
            big_ent_dens += 1
        else:
            ents_per_sent.append(len(ents))
            pass
        
        #ents_per_sent.append(len(ents))
        for ent in ents:
            if len(ent) > 10:
                big_ents += 1
                continue
            ent_lens.append(len(ent))
#             if len(ent) > 30:
#                 print(ent)
#                 for w, t in zip(words, tags):
#                     print(w, t)
    
    print("total sentences: ", len(sent_lens))
    print("total ents: ", len(ent_lens))
    
    print("mean sentence length: ", mean(sent_lens))
    print("max sentence length: ", max(sent_lens))
    print("mean entity density: ", mean(ents_per_sent))
    print("max entity density: ", max(ents_per_sent))
    print("ent density higher than 10: ", big_ent_dens)
    print("mean entity length: ", mean(ent_lens))
    print("max entity length: ", max(ent_lens))
    print("ent higher than 10: ", big_ents)
    print("sents higher than 120: ", big_sents)
    
    total_tags = counts['B'] + counts['I'] + counts['O']
    print("TAG COUNTS---")
    print("B: ", counts['B'], counts['B']/total_tags)
    print("I: ", counts['I'], counts['I']/total_tags)
    print("O: ", counts['O'], counts['O']/total_tags)
    
    return sent_lens, ent_lens, ents_per_sent
```

```python
os.listdir('..')
```

```python
filepaths = []

print("biored ----")
filepath = '../corpus/biored/biored_train_partial_all.csv'
biored_sent_lens, biored_ent_lens, biored_ent_dens = eda(filepath)

print("\nmatscholar ----")
filepath = '../corpus/matscholar/matscholar_train_bio_partial.csv'
matscholar_sent_lens, matscholar_ent_lens, matscholar_ent_dens = eda(filepath)

print("\nscierc ----")
filepath = '../corpus/scierc/scierc_train_partial.csv'
scierc_sent_lens, scierc_ent_lens, scierc_ent_dens = eda(filepath)

print("\nmedmentions ----")
filepath = '../corpus/medmentions/medmentions_small_bio.csv'
medmentions_sent_lens, medmentions_ent_lens, medmentions_ent_dens = eda(filepath)

print("\nmarpro ----")
filepath = '../corpus/matpro/matpro_bio_tags.csv'
matpro_sent_lens, matpro_ent_lens, matpro_ent_dens = eda(filepath)

print("\nscirex ----")
filepath = '../corpus/scirex/scirex_train_bio.csv'
scirex_sent_len, scirex_ent_len, scirex_ent_dens = eda(filepath)
```

## plotting sentence length histogram

```python
plt.rcParams["figure.autolayout"] = True
fig, ax = plt.subplots()
sns.histplot(biored_sent_lens, ax=ax, multiple='layer', color='dodgerblue', element='poly', label='BioRED', stat='probability')
sns.histplot(matscholar_sent_lens, ax=ax, multiple='layer', color='orchid', element='poly', label='MatScholar', stat='probability')
sns.histplot(scierc_sent_lens, ax=ax, multiple='layer', color='lightseagreen', element='poly', label='SciERC', stat='probability')
# plum or orchid for the last one
ax.set(xlabel='sentence length', ylabel='# sentences')
plt.legend(fontsize=20)
plt.show()
```

## plotting entity length histograms

```python
fig, ax = plt.subplots()
sns.histplot(biored_ent_lens, binwidth=1, ax=ax, multiple='layer', color='dodgerblue', label='BioRED')
sns.histplot(matscholar_ent_lens, binwidth=1, ax=ax, multiple='layer', color='orchid', label='MatScholar')
sns.histplot(scierc_ent_lens, ax=ax, binwidth=1, multiple='layer', color='lightseagreen', label='SciERC')
ax.set(xlabel='entity length', ylabel='# entities')
plt.legend(fontsize=20)
plt.show()
```

## plotting histogram for % of entities per entity length

```python
bins = range(1, 11)
fig, ax = plt.subplots()
ent_lens = [biored_ent_lens, matscholar_ent_lens, scierc_ent_lens]
labels = ['BioRED', 'MatScholar', 'SciERC']
colors = ['dodgerblue', 'orchid', 'lightseagreen']
plt.hist(ent_lens, bins=bins, label=labels, density=True, color=colors)
ax.set(xlabel='entity length', ylabel='fraction of entities')
plt.legend(fontsize=20)
plt.show()
```

## plotting histogram for entity density

```python
bins = range(1, 11)
fig, ax = plt.subplots()
ent_lens = [biored_ent_dens, matscholar_ent_dens, scierc_ent_dens]
labels = ['BioRED', 'MatScholar', 'SciERC']
colors = ['dodgerblue', 'orchid', 'lightseagreen']
plt.hist(ent_lens, bins=bins, label=labels, density=True, color=colors)
ax.set(xlabel='entity density', ylabel='fraction of sentences')
plt.legend(fontsize=20)
plt.show()
```
