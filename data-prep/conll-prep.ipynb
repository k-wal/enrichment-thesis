```python
import pandas as pd
from tqdm import tqdm
```

```python
class ConllPrep():
    def __init__(self):
        pass
    
    def get_doc_df(self, text):
        rows = []
        sents = text.split('\n\n')

        # for each sentence, create sentence and word_labels
        for sent in sents:
            if sent.strip() == '':
                continue
            sentence, tags = [], []
            lines = sent.split('\n')
            for line in lines:
                if line.strip() == '':
                    continue
                parts = line.split(' ')
                sentence.append(parts[0])
                tags.append(parts[-1][0])
            rows.append({'sentence':' '.join(sentence), 'word_labels':','.join(tags)})
        df = pd.DataFrame(rows)
        df = df.dropna().reset_index()
        return df
    
    def get_dir_df(self, input_path, output_path):
        df = pd.DataFrame(columns=['sentence', 'word_labels'])
        
        f = open(input_path, 'r')
        text = f.read()
        docs = text.split('-DOCSTART- -X- -X- O\n\n')
        
        for doc in tqdm(docs):
            cur_df = self.get_doc_df(doc)
            df = pd.concat([df, cur_df], ignore_index = True)
        
        df.to_csv(output_path, header=True)
        return df
```

```python
input_path = '../corpus/conll/conllpp_train.txt'
output_path = '../corpus/conll/conll_bio.csv'
prepper = ConllPrep()
df = prepper.get_dir_df(input_path, output_path)
```
